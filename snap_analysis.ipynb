# Preprocessing

```python
# Adding day of week column and fixing data types 
in_2022['published_at'] = pd.to_datetime(in_2022['published_at'])
in_2022['day_name'] = in_2022['published_at'].dt.day_name()

in_yearly['published_at'] = pd.to_datetime(in_yearly['published_at'])
in_yearly['day_name'] = in_yearly['published_at'].dt.day_name()
```

```python
#Cut off tail averages to provide accuracy in inaugural data (i.e data collected in first 7 days of episode on platform - Mar08 = -7days from Mar15 cutoff)
updated_yearly = in_yearly.loc[(in_yearly['published_at'] >= '2021-09-01') & (in_yearly['published_at']<= '2022-03-09')] 
updated_2022 = in_2022.loc[in_2022['published_at']<= '2022-03-09']
```
